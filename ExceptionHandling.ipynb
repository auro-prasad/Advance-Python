```python
class Grade:
 def __init__(self):
  self.g={}
 def add(self,i,a):
  self.g[i]=a
 def update(self,i,a):
  self.g[i]=a
 def delete(self,i):
  del self.g[i]
 def show(self):
  print(self.g)

s=Grade()
while True:
 print("1 add 2 update 3 delete 4 show 5 exit")
 c=input()
 try:
  if c=="1":
   i=input()
   a=int(input())
   s.add(i,a)
  elif c=="2":
   i=input()
   a=int(input())
   s.update(i,a)
  elif c=="3":
   i=input()
   s.delete(i)
  elif c=="4":
   s.show()
  else:
   break
 except:
  print("error")
```

```python
class Contact:
 def __init__(self):
  self.c={}
 def add(self,n,p):
  if n in self.c:raise Exception
  if not p.isdigit():raise Exception
  self.c[n]=p
 def edit(self,n,p):
  self.c[n]=p
 def search(self,n):
  print(self.c[n])

b=Contact()
while True:
 print("1 add 2 edit 3 search 4 exit")
 x=input()
 try:
  if x=="1":
   n=input()
   p=input()
   b.add(n,p)
  elif x=="2":
   n=input()
   p=input()
   b.edit(n,p)
  elif x=="3":
   n=input()
   b.search(n)
  else:
   break
 except:
  print("error")
```

```python
class Bank:
 def __init__(self):
  self.a={"1001":5000,"1002":3000}
 def transfer(self,s,d,m):
  if s not in self.a or d not in self.a:raise Exception
  if self.a[s]<m:raise Exception
  self.a[s]-=m
  self.a[d]+=m
  print(self.a)

b=Bank()
try:
 s=input()
 d=input()
 m=int(input())
 b.transfer(s,d,m)
except:
 print("transaction error")
```

```python
class Shop:
 def __init__(self):
  self.s={"shirt":10,"shoes":5}
 def order(self,p,q,c):
  if p not in self.s:raise Exception
  if self.s[p]<q:raise Exception
  if c!="SAVE10":raise Exception
  self.s[p]-=q
  print("order placed")

x=Shop()
try:
 p=input()
 q=int(input())
 c=input()
 x.order(p,q,c)
except:
 print("order error")
```

```python
class OutOfStockError(Exception):
    pass
class InvalidProductIDError(Exception):
    pass
class Inventory:
    def __init__(self):
        self.products={1:["Laptop",5],2:["Mouse",10],3:["Keyboard",0]}
    def purchase(self,pid,qty):
        if pid not in self.products:
            raise InvalidProductIDError("Invalid Product ID")
        if self.products[pid][1]<qty:
            raise OutOfStockError("Out of Stock")
        self.products[pid][1]-=qty
        print("Purchased",self.products[pid][0])
inv=Inventory()
try:
    inv.purchase(3,1)
except OutOfStockError as e:
    print(e)
except InvalidProductIDError as e:
    print(e)
```

```python
class Flight:
 def __init__(self):
  self.seats=5
 def book(self,n,s):
  if n=="" :raise Exception
  if s>self.seats:raise Exception
  self.seats-=s
  print("booked")

f=Flight()
try:
 n=input()
 s=int(input())
 f.book(n,s)
except:
 print("booking failed")
```
